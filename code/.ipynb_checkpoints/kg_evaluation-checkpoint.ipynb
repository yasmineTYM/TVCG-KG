# 1. Statistics Evaluation based on RDF 

```python
from rdflib import Graph

# Create an RDFLib graph
g = Graph()

# Load TTL data from a file
g.parse("../kg_data/kg_v4.ttl", format="turtle")
```

```python
# Count nodes
query_count_nodes = """
SELECT (COUNT(DISTINCT ?node) AS ?nodeCount)
WHERE {
    ?node ?p ?o
}
"""

# Count triplets (relationships)
query_count_triplets = """
SELECT (COUNT(*) AS ?tripletCount)
WHERE {
    ?s ?p ?o
}
"""

# Execute the queries
node_count_results = g.query(query_count_nodes)
triplet_count_results = g.query(query_count_triplets)

# Retrieve and print the counts
for row in node_count_results:
    node_count = row["nodeCount"]
    print(f"Total Number of Nodes: {node_count.value}")

for row in triplet_count_results:
    triplet_count = row["tripletCount"]
    print(f"Total Number of Triplets (Relationships): {triplet_count.value}")
```

```python
# Query for the number of distinct entity types (classes)
entity_type_query = """
SELECT (COUNT(DISTINCT ?entityType) AS ?entityTypeCount)
WHERE {
    ?s a ?entityType .
}
"""

# Query for the number of distinct relation types (predicates)
relation_type_query = """
SELECT (COUNT(DISTINCT ?relationType) AS ?relationTypeCount)
WHERE {
    ?s ?relationType ?o .
}
"""

# Execute the entity type query
entity_type_results = g.query(entity_type_query)

# Execute the relation type query
relation_type_results = g.query(relation_type_query)

# Get and print the counts
for row in entity_type_results:
    entity_type_count = row["entityTypeCount"]
    print(f"Number of Entity Types (Classes): {entity_type_count.value}")

for row in relation_type_results:
    relation_type_count = row["relationTypeCount"]
    print(f"Number of Relation Types (Predicates): {relation_type_count.value}")
```

```python
# Query for the number of distinct relation types and unique relation types for each entity type
relation_types_query = """
SELECT ?entityType (COUNT(DISTINCT ?p) AS ?relationTypeCount) (GROUP_CONCAT(DISTINCT ?p; separator="|") AS ?uniqueRelationTypes)
WHERE {
  ?s a ?entityType .
  ?s ?p ?o .
}
GROUP BY ?entityType
"""

# Execute the query
results = g.query(relation_types_query)

# Print the results
total = []
for row in results:
    entity_type = row["entityType"]
    relation_type_count = row["relationTypeCount"]
    unique_relation_types = row["uniqueRelationTypes"]
    
    # Convert the uniqueRelationTypes to a list
    unique_relation_types_list = unique_relation_types.split("|")
    
    total.append(relation_type_count.value)
    print(f"Entity Type: {entity_type}")
    print(f"Number of Distinct Relation Types: {relation_type_count.value}")
    print(f"Unique Relation Types: {', '.join(unique_relation_types_list)}")
    print()

```

```python
sum(total)/len(total)
```

```python
# Query for the average in-degree
average_in_degree_query = """
SELECT (AVG(?inDegree) AS ?averageInDegree)
WHERE {
  ?s ?p ?o .
  {
    SELECT ?o (COUNT(*) AS ?inDegree)
    WHERE {
      ?s ?p ?o .
    }
    GROUP BY ?o
  }
}
"""
in_degree_results = g.query(average_in_degree_query)
```

```python
# len(in_degree_results)
```

```python
# Print the results
for row in in_degree_results:
    average_in_degree = row["averageInDegree"]
    print(f"Average In-Degree: {average_in_degree.value}")
```

```python
# Query for the average out-degree
average_in_degree_query = """
SELECT ?o (COUNT(*) AS ?inDegree)
WHERE {
  ?s ?p ?o .
}
GROUP BY ?o
"""

# Execute the out-degree query
results = g.query(average_in_degree_query)
in_degree_results=[]
for row in results:
    in_degree_results.append(int(row['inDegree']))
```

```python
print(f"Average in-Degree:{sum(in_degree_results)/len(in_degree_results)}")
```

```python
# Query for the average out-degree
average_out_degree_query = """
SELECT ?s (COUNT(*) AS ?outDegree)
WHERE {
  ?s ?p ?o .
}
GROUP BY ?s
"""

# Execute the out-degree query
results = g.query(average_out_degree_query)
out_degree_results=[]
for row in results:
    out_degree_results.append(int(row['outDegree']))
```

```python
print(f"Average Out-Degree:{sum(out_degree_results)/len(out_degree_results)}")
```

```python
from SPARQLWrapper import SPARQLWrapper
sparql_endpoint = "https://makg.org/sparql"
sparql = SPARQLWrapper(sparql_endpoint)  
sparql.setReturnFormat('json')
```

```python
def queryAuthor(author):
    query  = f"""
    PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
    PREFIX magc: <https://makg.org/class/>
    PREFIX dcterms: <http://purl.org/dc/terms/>
    PREFIX foaf: <http://xmlns.com/foaf/0.1/>
    PREFIX fabio: <http://purl.org/spar/fabio/>
    PREFIX prism: <http://prismstandard.org/namespaces/basic/2.0/>
    PREFIX xsd: <http://www.w3.org/2001/XMLSchema#>

    SELECT *
    WHERE {{
    ?author a magc:Author;
    foaf:name '{author}'^^xsd:string.
    }}
    
    """
    sparql.setQuery(query)
    results = sparql.query().convert()
    return results['results']['bindings']
```

```python
from rdflib import Graph, Namespace

# Create an RDFLib graph
g = Graph()
# Load TTL data from a file
g.parse("../kg_data/kg_v4.ttl", format="turtle")
q = """
SELECT ?authorName
WHERE{
    ?author a <https://makg.org/class/Author>.
    ?author <http://xmlns.com/foaf/0.1/name> ?authorName
}
"""
results = g.query(q)
all_authors = []
for row in results:
    all_authors.append(str(row['authorName']))
```

```python
matched = 0
print(len(all_authors))
matched_results = []
```

```python
for i in range(len(all_authors)):
    ele = all_authors[i]
    if i%10==0:
        print(i)
    try:
        rel = queryAuthor(ele)
        if len(rel)>0:
            matched_results.append(rel)
            matched+=1
    except:
        matched_results.append([])
```

```python
matched, len(matched_results)
```

```python
matched/len(all_authors)
```

### 2. Paper

```python
def queryPaper(doi):
    query  = f"""
    PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
    PREFIX magc: <https://makg.org/class/>
    PREFIX dcterms: <http://purl.org/dc/terms/>
    PREFIX foaf: <http://xmlns.com/foaf/0.1/>
    PREFIX fabio: <http://purl.org/spar/fabio/>
    PREFIX prism: <http://prismstandard.org/namespaces/basic/2.0/>
    PREFIX xsd: <http://www.w3.org/2001/XMLSchema#>
    PREFIX datacite: <http://purl.org/spar/datacite/>
    
    SELECT *
    WHERE {{
    ?paper a magc:Paper;
    datacite:doi '{doi}'^^xsd:string;
    dcterms:title ?title.
    }}
    """
    sparql.setQuery(query)
    results = sparql.query().convert()
    return results['results']['bindings']
```

```python
q = """
SELECT ?paper
WHERE{
    ?paper a <https://makg.org/class/Paper>.
}
"""
results = g.query(q)
all_paper_dois = []
for row in results:
    all_paper_dois.append(str(row['paper']))
```

```python
print(len(all_paper_dois))
matched=0
for i in range(len(all_paper_dois)):
    ele = all_paper_dois[i]
    if i%10==0:
        print(i)
    try:
        rel = queryPaper(ele.replace('https://doi.org/',''))
        if len(rel)>0:
            matched+=1
        else:
            matched+=0
    except:
        matched+=0
```
